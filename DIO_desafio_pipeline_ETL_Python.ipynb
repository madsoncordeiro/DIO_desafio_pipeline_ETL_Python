### 2. Modificando os tipos de dados das colunas do dataframe

```python
# Modificando os tipos das colunas de float64 para int64 e, depois, para string
df['cpf'] = df['cpf'].fillna(0).astype('int64').astype('string')
df['ideCadastro'] = df['ideCadastro'].fillna(0).astype('int64').astype('string')
df['nuCarteiraParlamentar'] = df['nuCarteiraParlamentar'].fillna(0).astype('int64').astype('string')
df['numRessarcimento'] = df['numRessarcimento'].fillna(0).astype('int64').astype('string')

# Modificando os tipos das colunas de object para string
df['txNomeParlamentar'] = df['txNomeParlamentar'].astype('string')
df['sgUF'] = df['sgUF'].astype('string')
df['sgPartido'] = df['sgPartido'].astype('string')
df['txtDescricao'] = df['txtDescricao'].astype('string')
df['txtDescricaoEspecificacao'] = df['txtDescricaoEspecificacao'].astype('string')
df['txtFornecedor'] = df['txtFornecedor'].astype('string')
df['txtCNPJCPF'] = df['txtCNPJCPF'].astype('string')
df['txtNumero'] = df['txtNumero'].astype('string')
df['txtPassageiro'] = df['txtPassageiro'].astype('string')
df['txtTrecho'] = df['txtTrecho'].astype('string')
df['urlDocumento'] = df['urlDocumento'].astype('string')

# Modificando os tipos das colunas de int64 para string
df['codLegislatura'] = df['codLegislatura'].astype('string')
df['numSubCota'] = df['numSubCota'].astype('string')
df['numEspecificacaoSubCota'] = df['numEspecificacaoSubCota'].astype('string')
df['indTipoDocumento'] = df['indTipoDocumento'].astype('string')
df['numLote'] = df['numLote'].astype('string')
df['nuDeputadoId'] = df['nuDeputadoId'].astype('string')
df['ideDocumento'] = df['ideDocumento'].astype('string')

# Modificando os tipos das colunas de object para datetime
df['datEmissao'] = pd.to_datetime(df['datEmissao'])
df['datPagamentoRestituicao'] = pd.to_datetime(df['datPagamentoRestituicao'])
```

### Inserir dados no banco SQL

```python
# Query para criar tabela
query_criar_tabela = """
CREATE TABLE IF NOT EXISTS despesas_pela_cota_para_exercicio_da_atividade_parlamentar (
    txNomeParlamentar VARCHAR(100),
    cpf VARCHAR(11),
    ideCadastro VARCHAR(6),
    nuCarteiraParlamentar VARCHAR(3),
    nuLegislatura INT(4),
    sgUF VARCHAR(2),
    sgPartido VARCHAR(25),
    codLegislatura VARCHAR(2),
    numSubCota VARCHAR(3),
    txtDescricao VARCHAR(120),
    numEspecificacaoSubCota VARCHAR(1),
    txtDescricaoEspecificacao VARCHAR(50),
    txtFornecedor VARCHAR(300),
    txtCNPJCPF VARCHAR(19),
    txtNumero VARCHAR(50),
    indTipoDocumento VARCHAR(1),
    datEmissao DATE,
    vlrDocumento FLOAT,
    vlrGlosa FLOAT,
    vlrLiquido FLOAT,
    numMes INT(2),
    numAno INT(4),
    numParcela INT(3),
    txtPassageiro VARCHAR(150),
    txtTrecho VARCHAR(60),
    numLote VARCHAR(15),
    numRessarcimento VARCHAR(1),
    datPagamentoRestituicao DATE,
    vlrRestituicao FLOAT,
    nuDeputadoId VARCHAR(10),
    ideDocumento VARCHAR(15),
    urlDocumento VARCHAR(250),
    dia_datEmissao INT(2),
    mes_datEmissao INT(2),
    ano_datEmissao INT(4)
)
"""
```

```python
# Executando a query de criação da tabela
try:
    engine.raw_connection().cursor().execute(query_criar_tabela)
except:
    print("Essa tabela já existe. ")
```

```python
# Query para inserir os dados na tabela criada
query_insercao_dados = """
INSERT INTO despesas_pela_cota_para_exercicio_da_atividade_parlamentar (
    txNomeParlamentar,
    cpf,
    ideCadastro,
    nuCarteiraParlamentar,
    nuLegislatura,
    sgUF,
    sgPartido,
    codLegislatura,
    numSubCota,
    txtDescricao,
    numEspecificacaoSubCota,
    txtDescricaoEspecificacao,
    txtFornecedor,
    txtCNPJCPF,
    txtNumero,
    indTipoDocumento,
    datEmissao,
    vlrDocumento,
    vlrGlosa,
    vlrLiquido,
    numMes,
    numAno,
    numParcela,
    txtPassageiro,
    txtTrecho,
    numLote,
    numRessarcimento,
    datPagamentoRestituicao,
    vlrRestituicao,
    nuDeputadoId,
    ideDocumento,
    urlDocumento,
    dia_datEmissao,
    mes_datEmissao,
    ano_datEmissao
)
VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
"""

# Inserindo os dados da origem (df) dentro da tabela no banco de dados do SQLAlchemy
registros_inseridos = df.tail(100).to_sql( # Inserindo os últimos 100 registros do dataframe
    'despesas_pela_cota_para_exercicio_da_atividade_parlamentar',
    engine,
    if_exists = 'append', # Se a tabela existir, serão inseridos os dados nela
    index = False, # Para não salvar o índice do dataframe Pandas como uma coluna no banco de dados
    method = 'multi' # Para passar múltiplos valores em uma única inserção de dados
    )

try:
    if registros_inseridos:
        print(f"Os dados de {registros_inseridos} registros foram inseridos com sucesso no banco de dados SQLAlchemy. ")
    else:
        print("Houve erro ao inserir os dados no banco de dados SQLAlchemy. ")
except:
    print(ValueError)
```
